# User statistics

The first step is to choose a user to analyse.  I'm going to choose my own user, buy you're free to change it to any user you want. We'll also load the  lastfm API object.

```python
import download_data as dd
import matplotlib.pyplot as plt

user = 'gustavog1'
last = dd.generate_api_object()
```

Now, we'll see scrobbles activity:

```python
t = last.get_user(user).get_recent_tracks(limit = 100)
```

```python
from collections import Counter

albums = []
artists = []
tracks = []
for i in t:
    tracks.append(str(i.track))
    artists.append(str(i.track.artist))
    albums.append(i.album + " - "+str(i.track.artist))
    
artists = dict(Counter(artists).most_common(10))

plt.rcParams.update({'font.size': 46})
plt.figure(figsize=(100,30))
plt.bar(range(len(artists)), list(artists.values()), align='center')
plt.xticks(range(len(artists)), list(artists.keys()))
plt.show()
```
